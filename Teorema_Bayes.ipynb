### ***Teorema de Bayes***
p_blanca = 4/10

p(p_blanca|1) + p(p_blanca|2) + p(p_blanca|3)

1             + 1             + 2

p(blanco) = p(p_blanca|1*p(1)) + p(p_blanca|2)*p(2) + p(p_blanca|3)*p(3)
          = 1/4 * 4/10          + 1/3 * 3/10          + 2/3 * 3/10
          = **0.4**
